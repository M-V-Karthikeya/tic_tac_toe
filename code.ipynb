```python
M = {1:' ' , 2:' ', 3:' ' , 4:' ' , 5:' ' , 6:' ' , 7:' ' , 8:' ' , 9:' '} # representin the board as a dictionary where keys are positions and their values are 'x' or 'o'
p = ['x','o']                    # initially all positions are empty and p is a list of players
def print_board(M):                  # fuction to print the board
    print(M[1]+'|'+M[2]+'|'+M[3])
    print('-+-+-')
    print(M[4]+'|'+M[5]+'|'+M[6])
    print('-+-+-')
    print(M[7]+'|'+M[8]+'|'+M[9])
    print("\n")
def over(M):                        # fuction which returns 1 if all the cells in the board are filled else it returns 0
    for i in list(M.keys()):
        if(M[i]==' '):
            return 0
    return 1
def reward(M):               # reward function which returns 1 if player x wins, and returns -1 if o wins, returns 0 otherwise 
    if(M[1]==M[2]==M[3]):     # if first row is same 
        if(M[1]==p[0]):        # and it is marked x then return 1 cause x has won
            return 1
        elif(M[1]==p[1]):       # if it is marked o then returns -1 cause o has won
            return -1
    if(M[4]==M[5]==M[6]):      # similarly for 2nd row
        if(M[4]==p[0]):
            return 1
        elif(M[4]==p[1]):
            return -1
    if(M[7]==M[8]==M[9]):       # 3rd row
        if(M[7]==p[0]):
            return 1
        elif(M[7]==p[1]):
            return -1
    if(M[1]==M[4]==M[7]):     # 1st column
        if(M[1]==p[0]):
            return 1
        elif(M[1]==p[1]):
            return -1
    if(M[2]==M[5]==M[8]):    # 2nd column
        if(M[2]==p[0]):
            return 1
        elif(M[2]==p[1]):
            return -1
    if(M[3]==M[6]==M[9]):    # 3rd column
        if(M[3]==p[0]):
            return 1
        elif(M[3]==p[1]):
            return -1
    if(M[1]==M[5]==M[9]):    # right diagoal
        if(M[1]==p[0]):
            return 1
        elif(M[1]==p[1]):
            return -1
    if(M[7]==M[5]==M[3]):     # left diagonal
        if(M[7]==p[0]):
            return 1
        elif(M[7]==p[1]):
            return -1
    return 0               # returning 0 in all other cases
def minimax(M,m):          # definfing minimax function where m is an indicator which tells if the player is maximizing player or not. minimax returns the reward and the optimal position to play                     
    r = reward(M)         # first we calculate reward for every state
    if(r==1):          # if r = 1 then it has reached the terminal state so we can return 1 
        return 1,0
    if(r==-1):          # if r = -1 we return -1
        return -1,0
    if(r==0 and over(M)):    # if r=0 and board is filled completely then it is a draw so we return 0
        return 0,0
    else:                  
        pos = 0          # first we assume position pos to be 0
        if(m==1):               # when maximizing player is playing we fix best score to -infinte at the start 
            best = -100000
            d = M.copy()
            for i in list(d.keys()):
                if(d[i]==' '):                  # we check reward obtained for all possible moves and return the best among all of them
                    d[i] = p[0]
                    s = minimax(d,0)[0]
                    d[i]=' '
                    if(s>best):
                        best = s
                        pos = i
            return best,pos
        else:
            best = 100000                    # when minimizing player is making the move we set the best to infinte
            d = M.copy()
            for i in list(d.keys()):
                if(d[i]==' '):               # check all possible cells and return the best with respect to minimizing player
                    d[i] = p[1]
                    s = minimax(d,1)[0]
                    d[i]=' '
                    if(s<best):
                        best = s
                        pos = i
            return best,pos
def o_turn():                                  # function for o's turn
    pos = int(input("enter position"))                # taking position as input and placing o over there
    if(M[pos]==' '):
        M[pos]=p[1]
    else:
        pos = int(input("enter another position"))
        M[pos]=p[1]
    print("O's turn:\n")
    print_board(M)
def x_turn():                                # fuction for x's turn
    d = M.copy() 
    pos = minimax(d,1)[1]              # calling minimax function setting m to 1 as x is maximizing player
    M[pos]=p[0]                            # updating that position with x in board M
    print("X's turn:\n")
    print_board(M)
def play(M):
    print_board(M)
    while(not over(M)):             # while board is empty we keep on calling o_turn and x_turn alternatively until some one wins
        o_turn()
        if(reward(M) == -1):
            print("o won")
            break
        if(over(M)):
            print("draw")
            break
        x_turn()
        if(reward(M) == 1):
            print("x won")
            break 
        if(over(M)):                    # if no one wins and all the cells are filled then we print draw
            print("draw")
            break
```
